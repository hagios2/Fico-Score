##### Multivariet Linear Regression

```python
import statsmodels.api as sm
```

```python
df = pd.read_csv('loan.csv')
```

```python
in_rate = df['Interest.Rate']
fico = df['FICO.Score']
l_amount = df['Loan.Amount']
```

```python
y = np.matrix(in_rate).transpose()
```

```python
x1 = np.matrix(fico).transpose()
x2 = np.matrix(l_amount).transpose() #reshape variables into columns
```

```python
x = np.column_stack([x1, x2]) #put variable into 1 2d matrix
```

```python
x3 = sm.add_constant(x) # add column to input data
```

```python
model = sm.OLS(y, x3) #create a model with ordinary least square 
```

```python
model_fit = model.fit()# train model
```

```python
print("The p values are ", model_fit.pvalues) #lies btn 0 - 0.5 but its 
#better when closer to 0
```

```python
print("The R squared value is: ", model_fit.rsquared) #lies -1 to 1
# its better when close to 1
```
